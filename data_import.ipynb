```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
city_bike_data0 = pd.read_csv("/Users/shardendujha/Documents/Solita_Project/2021-05.csv")
city_bike_data1 = pd.read_csv("/Users/shardendujha/Documents/Solita_Project/2021-06.csv")
city_bike_data2 = pd.read_csv("/Users/shardendujha/Documents/Solita_Project/2021-07.csv")
```

```python
merged_df = pd.concat([city_bike_data0, city_bike_data1, city_bike_data2])
```

```python
merged_df
```

```python
# check data type
print(merged_df.dtypes)
```

```python
# Check for missing values
print(merged_df.isna().sum())
```

```python
merged_df = merged_df.dropna()
print(merged_df.isna().sum())
```

```python
# Check for duplicate rows
print(merged_df.duplicated().sum())

# Drop duplicate rows
merged_df = merged_df.drop_duplicates()
```

```python
merged_df
```

```python
# Don't import journeys that lasted for less than ten seconds
df_gt_10_s = merged_df.loc[merged_df['Duration (sec.)'] >= 10]
```

```python
df_gt_10_s
```

```python
# Don't import journeys that covered distances shorter than 10 meters
df_gt_10_m = df_gt_10_s.loc[df_gt_10_s['Covered distance (m)'] >= 10]        
```

```python
df_gt_10_m
```

```python
filtered_data = df_gt_10_m
```

```python
filtered_data.to_csv('filtered_data.csv', index=False)
```

```python
final_merged_data = pd.read_csv("filtered_data.csv")
```
